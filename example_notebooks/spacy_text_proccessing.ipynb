## Замена на гласные, согласные, пробелы и все остальное

```python
import re

vowel_re = re.compile(r"[ауоыэияюеёАУОЫЭИЯЮЕЁ]")
consonant_re = re.compile(r"[бвгджзйклмнпрстфхцчшщБВГДЖЗЙКЛМНПРСТФХЦЧШЩ]")
space_re = re.compile(r" ")

def transform_symbols(symbol):
    symbol = vowel_re.sub("v", symbol)
    symbol = consonant_re.sub("c", symbol)
    symbol = space_re.sub("s", symbol)
    symbol = re.sub(r"[^vcs]", "o", symbol)
    return symbol
```

```python
with open("/home/team/text_input/DD_txt/Dasha_Vasilieva_txt/DontsovaDaria_Dasha_Vasilieva_00.txt", "r", encoding="utf-8") as infile, \
     open("DontsovaDaria_Dasha_Vasilieva_00_trans.txt", "w", encoding="utf-8") as outfile:
    
    for line in infile:
        transformed_line = transform_symbols(line)
        outfile.write(transformed_line)

print("Done")
```

## Keep only lemmas and punct

```python
import spacy
nlp = spacy.load("ru_core_news_sm")

def simplify_line(line):
    doc = nlp(line)
    # keep lemmas, do not skip punctuation and prepositions
    string = ''
    for token in doc:
        if token.is_punct:
            string = string + token.text + ' '
        else:
            string = string + token.pos_ + ' '
    return string

def simplify_file_poc_punc(direc):
    input_file = Path("/home/team/text_input/DD_txt") / direc
    output_file = Path("DD_txt_pos_punc") / direc

    output_file.parent.mkdir(parents=True, exist_ok=True)

    with open(input_file, "r", encoding="utf-8") as infile, \
         open(output_file, "w", encoding="utf-8") as outfile:

        for line in infile:
            if line.strip():
                simplified = simplify_line(line)
                outfile.write(simplified + "\n")

    # print("Done")
```

```python
from multiprocessing import Pool

with Pool(29) as p:
    print(p.map(simplify_file_poc_punc, files))
```

- ADJ: adjective (a)
- ADP: adposition (b)
- ADV: adverb (c)
- AUX: auxiliary verb (d)
- CCONJ: coordinating conjunction (e)
- DET: determiner (f)
- INTJ: interjection (g)
- NOUN: noun (h)
- NUM: numeral (i)
- PART: particle (j)
- PRON: pronoun (k)
- PROPN: proper noun (h)
- PUNCT: punctuation
- SCONJ: subordinating conjunction (m)
- SYM: symbol (n)
- VERB: verb (o)
- X: other (p)

```python
from pathlib import Path
import spacy
nlp = spacy.load("ru_core_news_sm")

def simplify_file_poc_punc_short(direc):
    input_file = Path("/home/team/spacy_text_proc/DD_txt_pos_punc") / direc
    text = input_file.read_text(encoding="utf-8")

    output_file = Path("DD_txt_pos_punc_short") / direc
    output_file.parent.mkdir(parents=True, exist_ok=True)

    text = text.replace('ADJ', 'a')
    text = text.replace('ADP', 'b')
    text = text.replace('ADV', 'c')
    text = text.replace('AUX', 'd')
    text = text.replace('CCONJ', 'e')
    text = text.replace('DET', 'f')
    text = text.replace('INTJ', 'g')
    text = text.replace('NOUN', 'h')
    text = text.replace('NUM', 'i')
    text = text.replace('PART', 'j')
    text = text.replace('PRON', 'k')
    text = text.replace('PROPN', 'h')
    text = text.replace('SCONJ', 'm')
    text = text.replace('SYM', 'n')
    text = text.replace('VERB', 'o')
    text = text.replace('X', 'p')    

    text = text.replace('SPACE', '') 
    text = text.replace('PUNCT', '')
    text = text.replace(' ', '')
    text = text.replace('\n', 't')

    with open(output_file, "w", encoding="utf-8") as outfile:
        outfile.write(text)
    
    # print(set(text))
```

```python
from pathlib import Path
import os
from collections import Counter
import spacy
import pandas as pd

nlp = spacy.load("ru_core_news_sm")

def poc_comp(direc):
    input_file = Path("/home/team/text_input/DD_txt") / direc

    text = input_file.read_text(encoding="utf-8")
    doc = nlp(text)
    pos_counts = Counter(token.pos_ for token in doc if not token.is_punct)

    # df = pd.DataFrame.from_records(list(pos_counts.items()), columns=['item', os.path.basename(direc)])

    return pd.Series(pos_counts, name=os.path.basename(direc))
```
